# Comprehensions in Python

- List Compreshensions : It is a shorter syntax to create a new list using values of an existing list.



```python
l = [1,2,3,4,5]
l1 = []
for i in l:
    l1.append(i**2)
print(l1)
```

```python
[i**2 for i in l]
```

```python
a = [0,1,2,3]
b = [i+1 for i in a]
print(b)
```

```python
[i for i in l if i % 2 ==0 ]
```

```python
l1 = ['aws','nlp','ml','genai','data science masters']
```

```python
[i.upper() for i in l1]
```

```python
list((i**2 for i in l))
```

# Set Comprehensions

- It is a shorter synatx to create a new set using values of an existing set.

```python
a = {0,1,2,4,6}
b = {i ** 2 for i in a}
b
```

# Dict Comprehensions 

- It is a shorter synatx to create a new dictionary using values of an existing dictionary.

```python
a = {'Hello':'World','First':1}
{val:k for k,val in a.items()}
```

```python
d = {"key1":1 ,"key2":2, "key3":3,"key4":4}
```

```python
{k:v**2 for k,v in d.items()}
```

```python
d.items()
```

```python
d
```

```python
{k:v for k,v in d.items() if v > 1}
```
