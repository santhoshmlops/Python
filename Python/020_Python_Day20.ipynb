```python
l = [1,2,3,34,4,[2,3,4],'azure','cloud']
```

```python
l1_int = []
l2_str = []

for i in l:
    logging.info("we are iterating through our list and our local var is{}".format(l))
    
    if type(i) == list:
        logging.info("I am inside if statement and I am trying to check list type" + str(i))
        for j in i:
            logging.info("I am in another for loop for list inside element" + str(j))
            if type(j) == int:
                logging.info(" I am inside if statment")
                l1_int.append(j)
                
    elif type(i) == int:
            l1_int.append(i)
            
    else :
        if type(i) == str:
            l2_str.append(i)
            
logging.info("my final result for int is {l1} and str is {l2}".format(l1=l1_int , l2=l2_str))
```
