### Comparison Operators

Comparison operators are used to compare some value or operand and based on a condition , produce a Boolean. Python has six
comparison operators :

1. Less than (<)
2. Less than or Equal to (<=)
3. Greater than (>)
4. Greater than equal to (>=)
5. Equal to (==)
6. Not equal to (!=)

```python
a = 5
b = 10
result = a + b  #Here a and b are operands and + is the operator
print(result)
```

```python
# Take a variable
volume_ratio = 1.618

#condtion less than
print(volume_ratio<2)  #The volume ratio is lower than 3 , thus the output is True
print(volume_ratio<1)  # The volume ratio is greater than 1 , thus the output is False
```

```python
#condtion less than or equal to
print(volume_ratio<=2)
print(volume_ratio<=1)
print(volume_ratio<=1.618)
```

```python
#CONDTION greater than
volume_ratio = 1.618
print(volume_ratio>2)  
print(volume_ratio>1)
```

```python
# Condtion greater than equal to
volume_ratio = 1.618
print(volume_ratio>=2)
print(volume_ratio>=1)
print(volume_ratio>=1.618)
```

```python
# condtion equal to
print(volume_ratio==1.618)
print(volume_ratio==2)
```

```python
#condtion not equal to
print(volume_ratio!=2)
print(volume_ratio!=1.618)
```

# Else statment 

1. An else statement can be combined with an if statement.
2. An else statement conatins the block of code that executes if the condtional expression in the if statement resolves to 0 or False value.



```python
pi = 3.14
golden_ratio = 1.618

if pi<golden_ratio:
    print(f'The number of pi{pi} is greater than golden ratio{golden_ratio}')
    
else:
    print(f'The golden ratio of {golden_ratio} is lower than pi ratio {pi}')
    
print('Done!')
```

# Elif statement

1. This statment is used in conjunction with the if statement to add some other condition which is evaluated if the condition in if statement fails.




```python
age = 5

if age > 6 :
    print("You can go to primary school")
elif age == 5 :
    print("You should go to kinsergarden")
else :
    print("You are a baby")
```
